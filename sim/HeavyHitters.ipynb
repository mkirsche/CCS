```python
cumuContainment = []
cumuCoverage = []

fn = 'sorted.txt'
with open(fn) as f:
    for line in f:
        tokens = line.split()
        totcontained = int(tokens[4])
        totcoverage = int(tokens[5])
        cumuContainment.append(totcontained)
        cumuCoverage.append(totcoverage)

print(len(cumuContainment))
print(cumuContainment[0:10])
print(cumuCoverage[0:10])
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter(np.arange(len(cumuContainment)), cumuContainment, color = 'orange')
plt.title('Containment from Longest Reads')
plt.xlabel('Number of reads allowed to contain other reads')
plt.ylabel('Number of contained reads')
plt.xlim(0, 2000)
plt.show()
```

```python
plt.scatter(np.arange(len(cumuCoverage)), cumuCoverage, color = 'purple')
plt.title('Coverage from Longest Reads')
plt.xlabel('Number of reads')
plt.ylabel('Number of bases in the reference covered')
plt.show()
```
